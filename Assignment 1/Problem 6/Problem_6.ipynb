# Stock Market Problem
### Probability of losing Re 1 = p
### Probability of gaining Re 1 = 1-p
Irrespective of the amount involved in the game, the probability of losing,'p' or gaining,(1-p) Re 1 remains constant throughout the game which makes the game resemble to a Bernoulli experiment where the outcome of each day is one Bernoulli trial. 

- Outcome on each day are independent.
- Sequence of outcomes on 20 days constitutes a Bernoulli process
- To model the number of losses/gains in the span of 20 days, we could deploy the popular distribution "Binomial distribution" (discrete probability distribution of the number of successes in a sequence of n independent experiments).

#### $P(k) =$ $\binom{20}{k} p^{ k}(1-p)^{(20-k)}$ where k denotes the number of occurences of loss of Re 1 in 20 days

On 0th day, Amount = Rs.10

#### Let 'X' be the Random variable which denotes the final amount at the end of the game and can take values in the range [-10,30]

#### Probablility that I would have atleast Rs.10 = P(X >= 10) =  $\sum \limits _{k = 0} ^{10} P(k)$ = $P(10) + P(9) + ... + P(0)$
#### Symbolic answer in terms of "p"
#### P(X >= 10 ) = $\sum \limits _{k = 0} ^{10} \binom{20}{k} p^{ k}(1-p)^{(20-k)}$

```julia
using Plots;
f(p) = sum(binomial(20,k)*(p^k)*((1-p)^(20-k)) for k = 0:10);
```

```julia
atleast_10 = [f(p) for p = 0:0.01:1];
plot(0:0.01:1,atleast_10,ylabel = "Probability",xlabel = "p,  probability of losing Re1",legend = false)
```
